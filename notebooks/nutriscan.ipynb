```python
# Droping the fips,state and country column
df = df.drop(columns=['FIPS', 'State', 'County'],axis=1)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
# Convert percentage strings to floats
percentage_columns = [
        'Diabetes_Percentage_2013', 'Obesity_Percentage_2017',
        'Obesity_Percentage_2012', 'Diabetes_Percentage_2008'
    ]
for col in percentage_columns:
    df[col] = df[col].astype(str).str.rstrip('%').astype('float') / 100.0
```
